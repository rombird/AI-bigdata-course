```python
# TypedDict와 dict(일반적인 딕셔너리)의 차이점 예시
# TypedDict -> 타입 안정성, 코드 가독성, IDE 지원, 문서화
from typing import Dict, TypedDict

# 일반적인 파이썬 딕셔너리(dict)
# annotated -> 자료형 명시 키, 값 모두 str로
sample_dict:Dict[str, str] = {
    'name' : '푸바오',
    'age' : '5',
    'hobby' : 'Eating',
}
sample_dict
```

```python
print(type(sample_dict))
```

```python
# TypedDcit
class Panda(TypedDict):
    name: str
    age: int
    hobby: str

typed_dict: Panda = {
    'name': '푸바오',
    'age' : 5,
    'hobby' : 'Eating'
}
typed_dict
```

```python
# dict에서 age 키의 값을 수정, 추가
sample_dict['age'] = 10        # 처음 'age'의 값을 문자열로 지정했지만 다른 자료형이더라도 오류 X(문자열 -> 정수형)
sample_dict['new_field'] = '추가정보'   # 새로운 쌍 추가 가능
sample_dict # 우리 눈에는 같은 sample_dict로 보이지만 다른 위치정보를 가지고 있다고 생각
```

```python
# typedDict에 age 키의 값을 수정
typed_dict['age'] = 10
print(typed_dict)
print(type(typed_dict))
```
