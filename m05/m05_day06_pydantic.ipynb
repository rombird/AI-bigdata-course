```python
# typing 라이브러리 관련 예제
from typing import List, Dict, Optional

# Optional : 특정 타입이거나 None일 수도 있다는 것을 명시적으로 나타내는 힌트
def get_user_info(user_id: int) -> Optional[Dict[str, str]]:
    """ for문을 이용해 useri_id와 일치하는 사용자 정보를 찾아 바환하는 함수"""
    # 사용자 목록 -> 리스트[딕셔너리{}] (리스트 안에 딕셔너리)
    users: List[Dict[str, str]] = [
        {'id':'1', 'name' : '철수'},
        {'id':'2', 'name' : '영희'}
    ]

    # 리스트를 순회하면서 조건 확인
    for user in users:
        # user_id는 in, user["id"]는 str이므로 비교를 위해 형변환
        if user['id'] == str(user_id):
            return user  # 일치하면 해당사용자 반환
    
    # 반복문을 다 돌고도 못찾으면 None 반환
    return None

```

```python
print(get_user_info(1))
print(get_user_info(2))
print(get_user_info(3))
```

```python
# pydantic 라이브러리 관련 예제
from pydantic import BaseModel

class User(BaseModel):
    id: int
    name: str
    email: str

# 자동으로 타입 변환 및 검증
user = User(id="1", name="철수", email='test@example.com')
print(user) # id가 문자열이었지만 int로 변환됨
```

```python
type(user) # __main__.User
```

```python
type(user.id) # id를 문자열로 넣었지만 int로 변환
```

```python
type(user.name)
```

```python
# Python에서의 직렬화 예시 - json
import json

data = {'id':1, 'name':'홍길동'} # dict

# 직렬화 (dict -> json)
json_str = json.dumps(data)
print(json_str)
print(type(json_str))
```

```python
# 역직렬화 (json -> dict)
obj = json.loads(json_str)
print(obj)
print(type(obj))
```

```python
# pydantic + BaseModel
from langchain_core.pydantic_v1 import BaseModel

class User(BaseModel):
    id: int
    name: str

u = User(id=1, name='홍길동')
print(u)

# 직렬화
print(u.json())
print(u.dict())
```

```python
# 역직렬화
u2 = User.parse_raw('{"id":2, "name":"이몽룡"}')
print(u2)
```

```python
# Pydantic -> BaseModel, Field
from typing import Literal
from pydantic import BaseModel, Field

class Grade(BaseModel):
    binary_score: Literal['yes', 'no'] = Field(description="관련성 점수 'yes' 또는 'no'")
    # 타입 레벨에서 'yes' 또는 'no'만 허용

g1 = Grade(binary_score='yes')
print(g1)
```

```python
# 직렬화 json
g1.model_dump()
```

```python
Grade.model_json_schema()
```

```python
g2 = Grade(binary_score='no')
g2
```

```python
g2.model_dump()
```

```python
g3 = Grade(binary_score='YES') # 에러 발생 <- Input should be 'yes' or 'no'
g3
```
