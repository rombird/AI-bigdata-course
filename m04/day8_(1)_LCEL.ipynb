```python
from dotenv import load_dotenv
import os 
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
```

```python
# prompt를 PromptTemplate객체로 생성
prompt = PromptTemplate.from_template('{topic}에 대해 쉽게 설명해주세요.')
model = ChatOpenAI(model='gpt-4.1-nano', temperature=0.1) # 모델명, temperature 낮을수록 사실적

# LCEL 문법 사용해서 체인처럼 연결
chain = prompt | model
```

```python
# input 딕셔너리에 주제 설정
input = {'topic' : ' 인공지능 모델의 학습 원리'}

# invoke() 호출 - 딕셔너리 형태로 입력값 전달(키:값), invoke()함수 호출 시 입력값을 전달함
chain.invoke(input)
```

```python
response = chain.invoke(input)
response.content
```

```python

# stream() 호출 - 글자가 실시간으로 출력되는듯한 효과로 출력됨
answer = chain.stream(input)

# 스트림 출력
for token in answer:
    print(token.content, end='', flush=True) # flush : 자연스럽게 print되도록 print함수에서 설정
```

```python
# LLM의 응답을 문자열로 변환하는 파서 -> StrOutputParser
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
```

```python
# chain에 출력 파서를 추가 <- 서로 연결하여 체인을 구성
# 프롬프트 | 모델 | 출력파서 
chain = prompt | model | output_parser 
```

```python
input = {'topic':'빅데이터 분석가로 취업하는 방법'}
chain.invoke(input)
```

```python
# token : 
answer = chain.stream(input)
for token in answer:
    print(token,end='',flush=True)
```
