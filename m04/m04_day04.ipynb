```python
diarization = pipeline("/content/싼기타_비싼기타.mp3")

# 파일로 생성 -> rttm 파일
with open('싼기타_비싼기타.rttm', 'w') as rttm:
  diarization.write_rttm(rttm)
```

```python
# 판다스를 활용해 데이터프레임 형태로 저장
# 싼기타_비싼기타.rttm 파일을 보면 빈칸으로 나누어져 있음
# rttm 을 csv로 변환
import pandas as pd

rttm_path = '싼기타_비싼기타.rttm' # rttm 파일 경로

df_rttm = pd.read_csv(
    rttm_path,
    sep=" ", # 구분자는 공백
    header=None, # 헤더는 없음
    names=['type', 'file', 'chnl', 'start', 'duration', 'C1', 'C2', 'speaker_id', 'C3', 'C4']
    )
display(df_rttm)
```

```python
# 화자분리는 잘 되고있지만 한 화자의 발언이 여러행에 나누어 출력되고 있음
# -> 같은 화자가 계속 이야기하는 경우에 하나로 합쳐보기
# start : 발언이 시작된 시간, duration : 발언이 지속된 시간
# end : 발언이 끝난 시간 -> 추가
df_rttm['end'] = df_rttm['start'] + df_rttm['duration']
display(df_rttm)
```

```python
df_rttm['number'] = None # number컬럼 생성 -> None으로 초기화

# 번호를 매김 -> 기준점 -> 첫번째 행 -> number를 0으로 지정
# 데이터프레임.at[행, 열]

df_rttm.at[0, 'number'] = 0
display(df_rttm)
```

```python
# 회차 정보 매기기
for i in range(1, len(df_rttm)):
  # speaker_id가 앞의 것과 비교해서 다르면
  if df_rttm.at[i, 'speaker_id'] != df_rttm.at[i-1, 'speaker_id']:
    # 다른 화자가 나오면 number가 1을 증가
    df_rttm.at[i, 'number'] = df_rttm.at[i-1, 'number'] + 1
  else:
    df_rttm.at[i, 'number'] = df_rttm.at[i-1, 'number']

display(df_rttm.head(10))
```

```python
# 같은 화자끼리 묶어서 정리
df_rttm_grouped = df_rttm.groupby('number').agg(
    start = pd.NamedAgg(column='start', aggfunc='min'),
    end = pd.NamedAgg(column='end', aggfunc='max'),
    speaker_id = pd.NamedAgg(column='speaker_id', aggfunc='first')
)
display(df_rttm_grouped)
```

```python
# 발화 시간 추가(말하는 시간)
df_rttm_grouped['duration'] = df_rttm_grouped['end'] - df_rttm_grouped['start']

# 인덱스 제거
df_rttm_grouped = df_rttm_grouped.reset_index(drop=True)

display(df_rttm_grouped)
```

```python
# 화자를 분리한 결과를 csv파일로 내보내기
df_rttm_grouped.to_csv('싼기타_비싼기타_rttm.csv', sep=',', index=False)
```
