```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    temperature=0.1,
    model_name='gpt-4o-mini',
)
question = '대한민국의 수도는 어디일까요?'

print(f'[답변] : {llm.invoke(question)}')
```
