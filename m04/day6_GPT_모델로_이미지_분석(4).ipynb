# TTS로 영어 듣기 평가 문제 만들기
OpenAI api에는 문장을 음성파일로 변환하는 TTS(text-to-speeach)가 존재


```python
# 라이브러리 불러오기 및 api key 설정
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
api_key=os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)
```

```python
# 첫번째 TTS 테스트
response=client.audio.speech.create(
    model='tts-1-hd',
    voice='alloy',
    input='Hello world! This is a TTS test',
)
response.write_to_file('hello_world.mp3') # mp3로 파일 생성

# 재생
import IPython.display as ipd

ipd.Audio('hello_world.mp3') # 'alloy'의 목소리로 input 문장 음성파일로 변환
```

```python
# 목소리를 바꾸고 TTS 테스트
# create speech로 들어가면 다양한 목소리 확인 가능
voice = 'ash'
mp3_file = f'hello_world_{voice}.mp3'

response=client.audio.speech.create(
    model='tts-1-hd',
    voice=voice,
    input=f"Hello world! I'm {voice}. This is a TTS test"
)

response.write_to_file(mp3_file)

# 재생
ipd.Audio(mp3_file)
```

```python
# 5일차에 받은 json파일 읽어보기 - 영어 스크립트 json 파일 읽기
import json
with open('./images/image_quiz_eng.json', 'r', encoding='utf-8') as f:
    eng_dict = json.load(f) # load -> 파이썬으로 읽기 쉽게 바뀜

eng_dict # 확인
```

```python
# 여러 목소리를 추가한 영어스크립트 json 파일 읽기
voices = ['alloy', 'ash', 'coral', 'echo', 'fable', 'onyx', 'nova', 'sage', 'shimmer']
for q in eng_dict:
    no = q['no'] # q의 key=no를 no에 저장
    quiz = q['eng']
    quiz = quiz.replace('- (1)', '- One.\t') # (1)로 되어있으면 읽기 힘드니까 one으로 바꿔주기
    quiz = quiz.replace('- (2)', '- Two.\t')
    quiz = quiz.replace('- (3)', '- Three.\t')
    quiz = quiz.replace('- (4)', '- Four.\t')

    print(no, quiz)

    # 문제 번호에 따라 목소리 선택
    voice = voices[no % len(voices)] # 문제번호 8%9
    
    response=client.audio.speech.create(
        model='tts-1-hd',
        voice=voice,
        input=f'#{no}, {quiz}',
    )

    response.write_to_file(f'./audio/{no}.mp3')
```

```python
# 재생
ipd.Audio('./audio/1.mp3')
```
